```python
# 18. Write a Python program to find the median among three given numbers.
lst =list(map(int,input().split()))
if (len(lst)%2 == 0):
    even_median = (lst[len(lst)//2-1] + lst[len(lst)//2])/2
    print(even_median)
else :
    odd_median = lst[(len(lst)//2 )]
    print(odd_median)
```
